```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
```

```python
a = pd.read_csv('USA_cars_datasets.csv')
a.head()
```

```python
del a['vin']
del a['lot']
del a['Unnamed: 0']
del a['condition']
a.head()
```

```python
## Top Brands
top_brand = []
for i in a['brand'].unique():
    top_brand.append([i, len(a[a['brand']  == i])])

pd.DataFrame(top_brand,columns = ['brand','freq']).sort_values(by='freq', ascending=False).head().plot(x = 'brand', y = 'freq', kind = 'bar', figsize = (5,5))
```

```python
all_brand = []
for i in a['brand'].unique():
    all_brand.append([i, len(a[a['brand']  == i])])

pd.DataFrame(all_brand,columns = ['brand','freq']).sort_values(by='freq', ascending=False).plot(x = 'brand', y = 'freq', kind = 'bar', figsize = (20,5))
```

```python
## Top Popular cars
top_popular = []

for i in a['brand'].unique():
    top_popular.append([i , len(a[a['brand']  == i])])

pd.DataFrame(top_popular,columns = ['brand','freq']).sort_values(by='freq', ascending=False).plot(x = 'brand', y = 'freq', kind = 'bar', figsize = (20,5))
```

```python
## Year trend in Car Industry
year_trend = []
for i in a['year'].unique():
    year_trend.append([i , len(a[a['year']  == i])])

pd.DataFrame(year_trend,columns = ['year','freq']).sort_values(by='year', ascending = True).plot(x = 'year', y = 'freq', kind = 'line', figsize = (5,5))
```

```python
## Popular color in car industry
popular_color = []
for i in a['color'].unique():
    popular_color.append([i , len(a[a['color']  == i])])

pd.DataFrame(popular_color, columns = ['color','freq']).sort_values(by='freq', ascending = False).plot(x = 'color', y = 'freq', kind = 'bar', figsize = (20,5))
```

```python
## Brand having average maximum price
a.groupby('brand').mean()['price'].sort_values(ascending = False).plot.bar(figsize = (20,5))
```

```python
## State having average maximum price of cars
a.groupby('state').mean()['price'].sort_values(ascending = False).plot.bar(figsize = (20,5))
```

```python
# Car price trend yearwise
a.groupby('year').mean()['price'].sort_values(ascending = False).plot.bar(figsize = (20,5))
```
